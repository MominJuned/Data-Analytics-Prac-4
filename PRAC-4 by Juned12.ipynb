```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data from a CSV Repository
```

```python
np.random.seed(30)

data = {
    "Product_ID": range(1,301),
    "Category": np.random.choice(["Electronics","Clothing","Home","Sports"],300),
    "Price": np.random.randint(500,20000,300),
    "Quantity": np.random.randint(1,10,300),
    "Region": np.random.choice(["North","South","East","West"],300)
}

df = pd.DataFrame(data)

df.head()
```

```python
df.shape
```

```python
# Create Derived Column
```

```python
df["Revenue"] = df["Price"] * df["Quantity"]

df.head()
```

```python
# Basic Data Exploration
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Data Source Analysis
```

```python
category_sales = df.groupby("Category")["Revenue"].sum()

category_sales
```

#### Visualization

```python
# Category Revenue Bar Chart

plt.bar(category_sales.index, category_sales.values,color="k")
plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
# Region Based Analysis
```

```python
region_sales = df.groupby("Region")["Revenue"].sum()

region_sales
```

```python
# pie chart

plt.pie(region_sales.values,
        labels=region_sales.index,
        autopct='%1.1f%%')

plt.title("Revenue Contribution by Region")
plt.show()
```

```python
# Scatter Plot
```

```python
plt.scatter(df["Price"], df["Revenue"])
plt.title("Price vs Revenue Relationship")
plt.xlabel("Price")
plt.ylabel("Revenue")
plt.show()

# Higher priced products can generate higher revenue depending on quantity sold
```

```python
# Simulating API Data Source
```

```python
# API dataset (Application Programming Interface)

api_data = {
    "User": ["A","B","C","D","E"],
    "Platform": ["Twitter","Instagram","Facebook","Twitter","YouTube"],
    "Followers": [1200,3500,2700,1500,5000]
}

df_api = pd.DataFrame(api_data)

df_api
```

```python
plt.bar(df_api["Platform"], df_api["Followers"],color="y")
plt.title("Followers by Platform (API Data)")
plt.xlabel("Platform")
plt.ylabel("Followers")
plt.show()
```
